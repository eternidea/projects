```python
#changing the data type of specific columns

string1=[1,2,3,4,5,7,8]+[x for x in range(69,89)]+[x for x in range(149,169)]
boole1=[6]+[x for x in range(9,29)]+[x for x in range(89,109)]
```

```python
train.columns
```

```python
str_col1=[]
for x in string1:
    str_col1.append(train.columns[x])
print(str_col1)
```

```python
bool_col1=[]
for x in boole1:
    bool_col1.append(train.columns[x])
print(bool_col1)
```

```python
train[str_col1]=train[str_col1].astype('str')
```

```python
train[bool_col1]=train[bool_col1].astype('bool')
```

```python
train.info()
```

```python
string2=[1,2,3,4,6,7]+[x for x in range(69,88)]+[x for x in range(148,168)]
boole2=[5]+[x for x in range(8,28)]+[x for x in range(88,108)]
```

```python
str_col2=[]
for x in string2:
    str_col2.append(test.columns[x])
print(str_col2)
```

```python
bool_col2=[]
for x in boole2:
    bool_col2.append(test.columns[x])
print(bool_col2)
```

```python
test[str_col2]=test[str_col2].astype('str')
```

```python
test[bool_col2]=test[bool_col2].astype('bool')
```

```python
#Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
for x in str_col1:
    train[x] = le.fit_transform(train[x])
```

```python
for y in bool_col1:
    train[y] = train[y].replace({True:1, False:0})
```

```python
for x in str_col2:
    test[x] = le.fit_transform(test[x])
```

```python
for y in bool_col2:
    test[y] = test[y].replace({True:1, False:0})
```

```python
##splitting the data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
xtrain = train.drop(['id', 'target'],1)
ytrain = train.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(xtrain, ytrain, random_state=1)
```

```python
opt_tree = DecisionTreeClassifier(random_state = 0, criterion = 'entropy')

dt_params = {'max_depth': range(1,20)}

grid_tree = GridSearchCV(opt_tree, dt_params)

grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
grid_tree.best_score_
```

```python
from sklearn.metrics import accuracy_score

estimator = grid_tree.best_estimator_

pred = estimator.predict(X_test)
print('test data accuracy: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
